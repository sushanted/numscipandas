```python
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
```

### Legends

```python
X = np.linspace(-np.pi*4,np.pi*4,200)

fig,ax = plt.subplots()

ax.plot(X,2*np.sin(X),label='2sin')
ax.plot(X,np.cos(X),label='cos')

ax.legend(loc = 'lower left')
#ax.plot(X,np.tan(X))

plt.show()
```

### Annotations

```python
# simple annotations

A = np.linspace(-2*np.pi,2*np.pi,180)

fig,ax = plt.subplots()

ax.plot(5*np.cos(A),5*np.sin(A))

ax.scatter(0,0)

ax.annotate('center',(0,0))

ax.plot([-5,5],[5,5])

ax.annotate('tangent',(5,5))

plt.show()
```

```python
# Annotation with arrows

A = np.linspace(-2*np.pi,2*np.pi,180)

fig,ax = plt.subplots()

ax.plot(5*np.cos(A),5*np.sin(A))

ax.scatter(0,0)

ax.annotate('center',xy=(0,0),xytext=(2,2),arrowprops=dict(facecolor='orange', shrink=0.05))

ax.plot([-5,5],[5,5])

ax.annotate('tangent',xy=(5,5),xytext=(4,2),arrowprops=dict(facecolor='red', shrink=0.05))

ax.annotate('circle',xy=(-5,0),xytext=(-2,-2),arrowprops=dict(facecolor='red', shrink=0.05))

plt.show()
```
