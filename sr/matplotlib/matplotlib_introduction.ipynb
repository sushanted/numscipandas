### Different formats for the plots

```python
# different formats

format_strings = {
"solid line style":'-',
"dashed line style":'--',
"dash-dot line style":'-.',
"dotted line style":':',
"point marker":'.',
"pixel marker":',',
"circle marker":'o',
"triangle_down marker":'v',
"triangle_up marker":'^',
"triangle_left marker":'<',
"triangle_right marker":'>',
"tri_down marker":'1',
"tri_up marker":'2',
"tri_left marker":'3',
"tri_right marker":'4',
"square marker":'s',
"pentagon marker":'p',
"star marker":'*',
"hexagon1 marker":'h',
"hexagon2 marker":'H',
"plus marker":'+',
"x marker":'x',
"diamond marker":'D',
"thin_diamond marker":'d',
"vline marker":'|',
"hline marker":'_',
}

for name,char in format_strings.items():
    print(name," : ",char)
    plt.plot([-1,4,-2,3,2,-4,5,6,1,2,-4-3],char)
    plt.show()
```

### Multiple plots

```python
fix,ax = plt.subplots()

random_data = np.random.randint(-100,100,size=(10,10))

print(random_data)

days = np.arange(10)

mins = [random_data[i].min() for i in days]
maxes = [random_data[i].max() for i in days]
means = [random_data[i].mean() for i in days]

ax.plot(
    days,mins,'r',
    days,maxes,'g',
    days,means,'b',
)

plt.show()

```

### Range of axes

```python
X = [-400,-100,-9,-4,3,5,6,4,3,2,20,24,34,43,456,400,100]
Y = [-400,-200,5,-4,5,5,6,2,3,2,25,18,33,44,500,350,300]


def with_axis(axis):
    fix,ax = plt.subplots()

    ax.plot(X,Y)
    
    if axis : ax.axis(axis)

    plt.show()

with_axis([])

# Now fixing the range : kind of zoom in

with_axis([-50,100,-50,100])

with_axis([-10,50,-10,50])

with_axis([-10,10,-10,10])

with_axis([1,7.5,1,7.5])

```

### Mathematical functions

```python
DEFAULT = np.arange(10)
X = DEFAULT

def draw(X=DEFAULT,Y=DEFAULT,style="."):
    
    fig,ax = plt.subplots(figsize=(8,8))
    ax.plot(X,Y,style)
    plt.show()
    X = DEFAULT
    
```

```python
# y = x

draw(Y=X)
```

```python
# y = mx


draw(Y=5*X)
```

```python
# y = mx + c



draw(Y = 5*X + 12)
```

```python
# y = X*X


draw(Y=X*X,style='-')
```

```python
# y = a*X*X + b*X + c



draw(Y = 3*X*X + 5*X +23,style='-')
```

```python
# y = sin(X)

X=np.linspace(-2*np.pi,2*np.pi)

draw(X,Y=np.sin(X),style='-')
```

```python
# y = sqrt( r*r - X*X) : a circle 

X1 = np.arange(-5,5.1,0.1)

Y1 = np.sqrt(25 - X1*X1)

Y2 = -np.sqrt(25 - X1*X1)

draw(np.concatenate((X1,-X1)),Y=np.concatenate((Y1,Y2)),style="-")
```

```python
# x = r * cos(a) , y = r * sin(a) : pythagorous, a is angle from 0 to 360 (0 to 2*pi radians)

for side in (3,4,5,6,10,20,360):

    A = np.linspace(0,2*np.pi,side+1)

    draw(5*np.cos(A),5*np.sin(A),style="-")
```

### Filling

```python
fig,ax = plt.subplots()

X = np.arange(0,np.pi*4,0.1)

Y = np.sin(X)

ax.plot(X,Y)



# syntax : X, Y1, Y2 : region between Y1 and Y2 is filled, if Y1 or Y2 is scalar, the line Y = scalar is considered

ax.fill_between(X,Y,-1, color='green', alpha=.5) 

ax.fill_between(X,1,Y, color='red', alpha=.5)

plt.show()
```

```python
fig,ax = plt.subplots()

X = np.arange(0,np.pi*4,0.1)

Y = np.sin(X)

ax.plot(X,Y)

# Due to alpha the colors get mixed

# syntax : X, Y1, Y2 : region between Y1 and Y2 is filled, if Y1 or Y2 is scalar, the line Y = scalar is considered

ax.fill_between(X,Y,-1, color='green', alpha=.5) 

ax.fill_between(X,Y,0, color='red', alpha=.5) 

ax.fill_between(X,1,Y, color='blue', alpha=.5)

plt.show()
```
