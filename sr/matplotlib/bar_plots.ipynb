```python
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

### Simple bar plots
#### Difference between histogram and bar plots : The width of the rectangles has no mathematical meaning in bar plots, in histogram it is the bin size, distribution of the frequencies.

#### You are just defining the rectangles and their colors

#### x : array : x-co-ordinate of each bar
#### y : scalar : y-co-ordinate of each bar
#### width :  array : width of each bar
#### height : array : height of each bar
#### tick_label : array : label of each bar

```python
plt.bar(x=[1,2,3,4],height=np.random.randint(1,10,size=(4)))
plt.show()
```

```python
bars = plt.bar(
    x=[1,2,3,4],
    height=np.random.randint(2,10,size=(4)),
    color=('red','green','blue','yellow')
)

plt.show()
```

```python
# Another way to do the same, set individual bar values
bars = plt.bar(x=[1,2,3,4],height=np.random.randint(2,10,size=(4)))

bars[0].set_color('red')
bars[0].set_width(0.5)
bars[1].set_color('green')
bars[1].set_height(6) # Always constant
bars[2].set_color('blue')
bars[2].set_y(1) # start from 1 : floating in the air!
bars[3].set_color('yellow')
bars[3].set_x(3.5) # starting at 3.5

plt.show()
```

#### Bar plot with labels

```python
dates = ["21-Oct-18","17-Nov-18","1-Feb-19","14-Mar-19","15-Jun-19","4-Aug-19","6-Nov-19","25-Nov-19","28-Nov-19","12-Dec-19","5-Feb-20","19-Feb-20","21-Feb-20","7-Mar-20","30-Mar-20"]
weights = np.array([71.3,71.8,71.7,72.3,71.2,72.6,74.2,74.2,73.5,73.8,71.6,70.9,72.3,68.7,68.7])
low = 66
high = 76
```

```python

fig,ax = plt.subplots(figsize=(10,10))

ax.set_title("My Weights")



plt.bar(np.arange(len(dates)),(weights-low)*2,0.5) # last param is width of the bar
plt.xticks(np.arange(len(dates)),dates,rotation=90)
plt.yticks(np.arange((high-low)*2),np.arange(low,high,0.5))

plt.show()
```

#### Bar plot function 

```python
plt.subplots(figsize=(15,5))

plt.bar(x=np.arange(len(dates)),height=weights-66,width=0.5,color="green", align="center",
       tick_label=dates)
plt.xticks(rotation=90)
plt.yticks(np.arange(10),np.arange(66,77))
plt.show()
```

#### Vertical bar charts

```python
plt.subplots(figsize=(20,15))

# use y instead of x and exchange height and width values
plt.barh(y=np.arange(len(dates)),width=(weights-low)*4,height=0.5,color="green", align="center",
       tick_label=dates)

# first param is the position
# second param is label text to be displayed
plt.xticks(np.arange((high-low)*4),np.arange(low,high,0.25),rotation=90)
plt.show()
```

### Grouped bar charts : 
#### there is no special method to group, it's just our calculation to separate the individuals in the group
#### With grouped charts individual data is clearly displayed, larger value bars doesn't overlay smaller value bars

```python
fig,ax = plt.subplots(figsize=(25,10))

width = 0.15

for index,person_weights in enumerate((M,A,W,S,N)):
    plt.bar(
            x=np.arange(len(person_weights))+index*width,
            width=width,
            height=person_weights,
            tick_label=dates,
            label="MAWSN"[index]
           )
    
ax.legend(loc='best')
plt.xticks(np.arange(len(dates))+2*width,dates,rotation=90)
plt.show()
```
