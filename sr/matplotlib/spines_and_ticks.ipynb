```python
from matplotlib import pyplot as plt

import numpy as np

%matplotlib inline
```

### Spine properties

```python


A = np.linspace(0,2*np.pi,360)

X = 8*np.cos(A)

Y = 5*np.sin(A)

fig,ax = plt.subplots()

# set spines properties
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))


ax.plot(X,Y)

plt.show()
```

### Axis properties : ticks

```python
# ticks position

fig,ax = plt.subplots()

ax.xaxis.set_ticks_position('top')
ax.yaxis.set_ticks_position ('right')

ax.plot([1,2,3,1],[3,6,-5,3])

ax.scatter([1,2,3],[3,6,-5],s=50,color='green')

plt.show()
```

```python
# specific ticks

fig,ax = plt.subplots()

A = np.linspace(0,2*np.pi,360)

X = 10*np.cos(A)

Y = 10*np.sin(A)

# More precise ticks
ax.xaxis.set_ticks(np.arange(-10,10,1))
# Less precise ticks
ax.yaxis.set_ticks(np.linspace(-10,10,5))

ax.plot(X,Y)

ax.scatter([0,0,0,10,-10],[0,10,-10,0,0])

plt.show()
```

```python
# tick lables

fig,ax = plt.subplots()

# Note : we had to calculate them manually
ax.xaxis.set_ticks([-10,0,10]) # this is important to decide the position of ticks
ax.xaxis.set_ticklabels(['left','center','right']) # this just labels the ticks

ax.yaxis.set_ticks([-10,0,10])
ax.yaxis.set_ticklabels(['bottom','center','top'])


ax.plot([-10,-10,10,10,-10],[-10,10,10,-10,-10])

ax.scatter([-10,-10,-10,0,0,0,10,10,10],[-10,0,10,-10,0,10,-10,0,10])

plt.show()
```

### Locators

```python
# first manual ticks

fig,ax = plt.subplots()

waves = 6

start = -waves*np.pi
end = -1*start

X = np.linspace(start,end,200)

Y = np.sin(X)

ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position(('data',0))

ticks = np.linspace(start,end,waves*2+1)# so much calculation!
ax.xaxis.set_ticks(ticks) 

# LaTex notation for pi
# np.around get the proper rounded value, astype just does truncation like 0.999 = 0
labels = (str(i)+r'$\pi$' for i in np.around(ticks/np.pi).astype(int))
ax.xaxis.set_ticklabels(labels)

plt.plot(X,Y)

plt.show()
```

```python
# ticks with locator

fig,ax = plt.subplots()

waves = 4

ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position(('data',0))

X = np.linspace(-np.pi*waves,np.pi*waves,200)

Y = np.sin(X)

# Automatic adjust the ticks distanced at np.pi from each other
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi))

ax.plot(X,Y)

plt.show()
```

```python
# Modify tick labels

fig,ax = plt.subplots()

waves = 2

X = np.linspace(-np.pi*waves,np.pi*waves,200)

Y = np.sin(X)

ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi/2))

ax.set_xticklabels([r'$-2\pi$', r'$-\frac{3\pi}{2}$', r'$-\pi$', 
                    r'$-\frac{\pi}{2}$', 0, r'$\frac{\pi}{2}$', 
                    r'$+\pi$', r'$\frac{3\pi}{2}$', r'$+2\pi$',r'$\frac{5\pi}{2}$'])

for xtick in ax.get_xticklabels():
    xtick.set_fontsize(18)

ax.plot(X,Y)

plt.show()
```
