#### Important Note about axes and subplots:

Copied from : https://stackoverflow.com/a/45053269

In English language the singular is axis and the plural is axes. Two of the kind axis form two axes.

In matplotlib, a matplotlib.axes._axes.Axes object is often simply called "axes". This object incorporates an xaxis and a yaxis, thus the name. But speaking of that object, one would call it axes in singular. Several of those are still called axes.

Every subplot is an Axes object, but there are Axes objects, which are no AxesSubplot object. E.g. an axes, which is created through the subplot mechanism is a matplotlib.axes._subplots.AxesSubplot. This class derives from matplotlib.axes._axes.Axes, thus this subplot is an axes. You can however also create axes via different mechanisms, e.g. by adding an axes to the figure, fig.add_axes(). This would then not be a subplot, but an axes, matplotlib.axes._axes.Axes

### Polar plotting

```python
# Instead of x and y, we use a and r, where a is angle and r is radius

fig,ax = plt.subplots(1,1,subplot_kw=dict(polar=True))

# first array is of angles and second is of radius
ax.plot([np.pi/2,np.pi/2],[4,10])

X = np.linspace(0,2*np.pi,180) # Cover all the possible angles

Y = X*0 + 4 # X*0 to get an array of same size as X, +4 to have all elements as 4 (radius = 4)

ax.plot(X,Y)

Y = 6 + np.random.random(X.size)

ax.plot(X,Y)

X = np.linspace(0,2*np.pi,9) # 9 points : first and last point same : 0 and 2pi

ax.plot(X,X*0+8)

plt.show()
```

### Creating complex plots

```python
def draw_plots(*plots):
    fig = plt.figure()
    
    for plot in plots:
        sub = plt.subplot(*plot)
        sub.set_xticks(())
        sub.set_yticks(())
        sub.text(0.5,0.5,f"{plot[0]} x {plot[1]} : {plot[2]}",ha='center')
    
    fig.tight_layout() # adjust the layout 
    plt.show()
```

```python
draw_plots((2,2,1),(2,2,4)) # The grid is 2x2, draw only first and fourth plots
```

```python
draw_plots((3,3,3),(3,3,5),(3,3,7)) # The grid is 3x3, draw only plots 3,5,7,
#other plots are not defined but their places are reserved
```

#### Combine cells using only joins

```python
draw_plots((2,2,(1,2)),(2,2,3),(2,2,4))
# Join first 2 plots into a single plot
```

```python
draw_plots((2,2,(1,3)),(2,2,2),(2,2,4))
# Join plots 1 and 3
```

```python
draw_plots((3,3,(1,2)),(3,3,(4,5)),(3,3,(7,8)),(3,3,(3,9)))
# Join plots 1,2 ; 4,5 ; 7;8 AND join plots 3 to 9 i.e. 3,6,9
```

```python
draw_plots((4,2,(1,7)),(4,2,2),(4,2,(4,6)),(4,2,8))
#  join plots 1 to 7 i.e. 1,3,5,7; join plots 4 and 6
```

```python
draw_plots((4,2,1),(4,2,2),(4,2,3),(4,2,(4,6)),(4,2,5),(4,2,(7,8)))
#  join plots 1 to 7 i.e. 1,3,5,7; join plots 4 and 6
```

#### Combine cells using plots of different grids

```python
draw_plots((2,1,1),(2,2,3),(2,2,4))
# (2,1,1) : this is 2x1 grid's first plot, we are not defining (2,1,2)
# instead we are fitting two plots of 2x2 grids in the place for (2,1,2)
# the 2x2 plots are indexed as 3 and 4 because these are their places in 2x2 grid
# the (2,1,1) has already taken the indices 1 and 2
```

```python
draw_plots((1,2,1),(2,2,2),(2,2,4))
# (1,2,1) : this is 1x2 grid's first(left) plot, we are not defining (1,2,2)
# instead we are fitting two plots of 2x2 grids in the place for (1,2,2)
# the 2x2 plots are indexed as 2 and 4 because these are their places in 2x2 grid
# the (1,2,2) has already taken the indices 1 and 3
```

```python
draw_plots((3,3,(1,2)),(3,3,(4,5)),(3,3,(7,8)),(1,3,3))
# Join plots 1,2 ; 4,5 ; 7;8 AND join plots 3 to 9 i.e. 3,6,9
```

### Creating complex plots using GridSpec

```python
import matplotlib.gridspec as gridspec 
```

```python
def draw_plots_gs(*grid_specs):
    fig = plt.figure(figsize=(8,8))
    
    for grid_spec in grid_specs:
        sub = plt.subplot(grid_spec)
        sub.set_xticks(())
        sub.set_yticks(())
        rows,columns,row_start,row_end,column_start,column_end = grid_spec.get_rows_columns()
        sub.text(0.5,0.5,f"{rows} x {columns}, row {row_start}:{row_end}, column {column_start}:{column_end}",ha='center')
    
    fig.tight_layout() # adjust the layout 
    plt.show()
```

```python
gs = gridspec.GridSpec(1,1)

draw_plots_gs(gs[0,0])
```

```python
gs = gridspec.GridSpec(2,2)

draw_plots_gs(gs[0,0],gs[1,1]) # define specific grids
```

```python
gs = gridspec.GridSpec(2,2)

draw_plots_gs(gs[0,1],gs[1,:]) # grid(0,1) and combine all from second row
```

```python
gs = gridspec.GridSpec(2,2)

draw_plots_gs(gs[:,0],gs[1,1]) # grid(1,1) and combine all from first column
```

```python
gs = gridspec.GridSpec(2,2)

draw_plots_gs(gs[:,:]) # combine all from both rows and both columns
```

```python
gs = gridspec.GridSpec(3,3)

draw_plots_gs(gs[0,:-1],gs[:,-1],gs[1,:-1],gs[2,:-1])
```

```python
gs = gridspec.GridSpec(4,2)

draw_plots_gs(gs[:,:-1],gs[0,-1],gs[1:-1,-1],gs[3,-1])
```

```python
gs = gridspec.GridSpec(4,2)

draw_plots_gs(gs[0,0],gs[0,-1],gs[1,0],gs[2,0],gs[1:-1,-1],gs[-1,:])
```

### Height and Width ratios to create subplots

```python
# Divide the width and height of the figure an given ratios (1:2:3 and 3:2:1)

ncols,nrows = 3,3

gs = gridspec.GridSpec(ncols, nrows, width_ratios=[1,2,3],
                          height_ratios=[3,2,1])

draw_plots_gs(*(gs[i,j] for i in range(ncols) for j in range(nrows)))

plt.show()
```

### Multilevel subplotting : sub-sub plots

```python
gs = gridspec.GridSpec(1,2)

gs1 = gs.subgridspec(3,2)
gs2 = gs.subgridspec(2,2)

draw_plots_gs(gs1[0,0],gs2[0,0])
```

```python
# Though it looks like one plot inside another, they are actually 2 axes added to the same figure

fig = plt.figure()

X = [1, 2, 3, 4, 5, 6, 7]
Y = [1, 3, 4, 2, 5, 8, 6]

axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9]) # main axes left,bottom,width,hight
axes2 = fig.add_axes([0.2, 0.6, 0.4, 0.3]) # inset axes left,bottom,width,hight

# main figure
axes1.plot(X, Y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('axes1')

# insert
axes2.plot(Y, X, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('axes2');
```

```python
fig, axes = plt.subplots(1, 4, figsize=(10, 4))

x = np.arange(0, 5, 0.25)

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

# There is no difference bertween following two, except signature
axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom ylim and xlim");

axes[3].plot(x, x**2, x, x**3)
axes[3].axis([2,5,0, 60])
axes[3].set_title("custom using axis");
```

### Multiple scales 
#### Create a twin Axes sharing the xaxis or yaxis

```python
# Two different functions, sharing xaxis but different yaxis values

fig, ax1 = plt.subplots()

x = np.arange(1,7,0.1)
ax1.plot(x, 2 * np.pi * x, lw=2, color="blue")
ax1.set_ylabel(r"Circumference $(cm)$", fontsize=16, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(x, np.pi * x ** 2, lw=2, color="darkgreen")
ax2.set_ylabel(r"area $(cm^2)$", fontsize=16, color="darkgreen")
for label in ax2.get_yticklabels():
    label.set_color("darkgreen")
    
ax3 = ax1.twiny()
#ax3.xaxis.set_ticks([x for x in range(0,8)])

plt.show()
```

```python
# Note : the same scale for differnt functions is not justifiable, in the following 
# example, the sqrt value is almost invisible (or seem to be 0)

fig,ax = plt.subplots()

X = np.arange(100)

ax.plot(X,np.square(X))
ax.plot(X,np.sqrt(X))

plt.show()
```

```python
# Solution for above problem:

fig,ax_sqrt = plt.subplots()

X = np.arange(10)

sqrt_color = 'green'
ax_sqrt.plot(X,np.sqrt(X),color=sqrt_color)
ax_sqrt.set_ylabel('Square',color=sqrt_color)
for tick in ax_sqrt.get_yticklabels():
    tick.set_color(sqrt_color)

sqr_color = 'red'
ax_sqr = ax_sqrt.twinx()
ax_sqr.plot(X,np.square(X),color=sqr_color)
ax_sqr.set_ylabel('Square Root',color=sqr_color)
for tick in ax_sqr.get_yticklabels():
    tick.set_color(sqr_color)


plt.show()
```
