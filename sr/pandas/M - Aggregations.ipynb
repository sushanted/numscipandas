```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randint(1,7,size=(8,8)))
df
```

```python
rollingWindow = df.rolling(window=3)
```

```python
rollingWindow.aggregate(np.sum) # this is same as rollingWindow.sum()
```

```python
rollingWindow[0].aggregate(np.sum) # same as rollingWindow[0].sum()
```

```python
rollingWindow[[0,2]].aggregate(np.sum) # same as rollingWindow[[0,2]].sum()
```

```python
# This is interesting : multiple functions on multiple columns
aggregatedFrame = rollingWindow[[0,2]].aggregate([np.sum,np.mean])
aggregatedFrame
```

```python
aggregatedFrame[0]
```

```python
aggregatedFrame[0]['sum']
```

```python
# operations per column
aggregatedFrame[[0,2]].aggregate({0:np.sum,2:np.mean})
```
