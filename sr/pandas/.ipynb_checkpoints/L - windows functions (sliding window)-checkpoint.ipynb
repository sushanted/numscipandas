```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randint(1,7,size=(8,8)))
df
```

```python
df.rolling(window=3).sum() 
# first 2 values are NaN because window is 3, at row 2 we get a complete window for the calculation
```

```python
# apply receives an array and a single value of same type is expected in return
df.rolling(window=3).apply(max)
```

```python
df.rolling(window=3).apply(lambda x : x[0]*x[1]+x[2])
```

```python
# cumilitive operation on all previous values
df.expanding().min()
```

```python
df.expanding().apply(max)
```

```python
# Array of all previous values is passed to apply
df.expanding().apply(len)
```

```python
# sum of squares of all values

df.expanding().apply(lambda arr : sum((x*x for x in arr)))
```

```python
# By default min_period is 1, we can change it
df.expanding(min_periods=4).apply(len) # now till we accumulate 4 elements, the apply will not be called
```

```python
# TODO : Learn it later

df.ewm(com=0.5).mean()
```
