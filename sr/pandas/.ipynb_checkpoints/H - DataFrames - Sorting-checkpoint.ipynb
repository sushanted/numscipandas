```python
import numpy as np
import pandas as pd
```

### Sorting can be done vertically (by index values) OR horizontally (by column values)

```python
df = pd.DataFrame(
    np.random.choice(['YES','NO','MAYBE'],(3,7)),
    columns=['S','M','T','W','T','F','S']
)
```

```python
# Sort by index values
df.sort_index(ascending=False)
```

```python
# Sort by column values
df.sort_index(axis=1)
```

```python
# sort by column value
df.sort_values(by='F')
```

```python
# sort by multiple column values
df.sort_values(by=['W','F'])
```

```python
# sort by multiple column values
df.sort_values(by=['W','F'],ascending=False)
```

```python
# sort by index values
df.sort_values(by=[0,2],ascending=False,axis=1)
```
